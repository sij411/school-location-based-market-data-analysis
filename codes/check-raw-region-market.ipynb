```python
def scatter_plot_for_service_freq(dt):
    """
        input: dt - data
        output: scatter chart with service counts
    """
    # 서비스 업종
    service = "서비스_업종_코드_명"

    dt[service].value_counts()

    # 서비스 업종과 빈도수 
    dt_service = dt[service].value_counts().reset_index()
    dt_service.columns = [service, "counts"]

    # 빈도수 통계 정보
    dt_service['counts'].describe()

    # sort

    dt_service_sorted = dt_service.sort_values(by='counts', ascending=False).head(10)

    # 빈도수가 높은 상위 10개 업종

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.scatter(dt_service_sorted[service], dt_service_sorted['counts'])

    ax.set_xlabel(service)
    ax.set_ylabel('counts')
    ax.set_title(f'빈도수가 높은 상위 10개 {service}')

    plt.show()
```

```python
# 전체 데이터 

scatter_plot_for_service_freq(dt)
```
